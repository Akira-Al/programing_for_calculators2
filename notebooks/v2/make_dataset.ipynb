```python
import pandas as pd
```

```python
train_text = pd.read_csv("../../data/v2/train_normalized.txt", sep="\t", header=None)
train_text.columns = ["text"]
train_label = pd.read_csv("../../data/v2/train_label.csv", sep="\t", header=None)
train_label.columns = ["label"]

val_text = pd.read_csv("../../data/v2/validation_normalized.txt", sep="\t", header=None)
val_text.columns = ["text"]
val_label = pd.read_csv("../../data/v2/val_label.csv", sep="\t", header=None)
val_label.columns = ["label"]
```

```python
train_df = pd.concat([train_text, train_label], axis=1)
val_df = pd.concat([val_text, val_label], axis=1)
```
